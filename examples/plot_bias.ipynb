```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
bias=-0.173
beta=1.58
alpha=2.9 # redshift evolution of bias
```

```python
# compute P3D at z=3, in the parallel (mu=0) and transverse (mu=1) directions
z=np.arange(2.0,4.1,0.1)
b=bias*((1+z)/(1+2.25))**alpha
```

```python
bias_file='../data/z_beta_bias.txt'
inz,inbeta,inbias = np.loadtxt(bias_file,unpack=True)
```

```python
plt.plot(z,bias*np.ones_like(z),label='reference')
plt.plot(z,b,label='python')
plt.plot(inz,-inbias,label='c++')
plt.legend()
plt.xlabel('z')
plt.ylabel(r'$b_\delta$')
plt.title('Linear bias as a function of redshift')
plt.savefig('bias_z.png')
```

```python
plt.plot(z,beta*np.ones_like(z),label='python')
plt.plot(inz,inbeta,label='c++')
plt.legend()
plt.xlabel('z')
plt.ylabel(r'$\beta$')
plt.title('Linear RSD beta as a function of redshift')
plt.savefig('beat_z.png')
```
